```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data = pd.read_csv("customer_data.csv")

df = pd.DataFrame(data)
df
```

```python
df_2 = df.head(100).copy()
df_2
```

```python
from sklearn.cluster import KMeans

x = df_2[["income", "spending"]]

plt.figure(figsize=(10, 6))
plt.scatter(x["income"], x["spending"], s=100, c="blue", edgecolors='k')
plt.title("Customer Distribution")
plt.xlabel("Annual Income")
plt.ylabel("Spending Score")
plt.grid(True)
plt.show()
```

```python
from sklearn.cluster import KMeans
KMeans = KMeans(n_clusters=3, random_state=0)
df_2['cluster'] = KMeans.fit_predict(x)
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df_2, x="income", y="spending", hue="cluster", palette="Set1", s=100, edgecolor='k')
centroids = KMeans.cluster_centers_
plt.scatter(centroids[:, 0], centroids[:, 1], s=300, c='black', marker='X', label='Centroids')
plt.title("Customer Segments K-Means Clusters")
plt.xlabel("Annual Income")
plt.ylabel("Spending Score")
plt.legend()
plt.grid(True)
plt.show()
```
